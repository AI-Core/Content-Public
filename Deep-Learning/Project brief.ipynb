# Deep Learning Project

## What?
Apply deep learning to your dataset to beat the baseline results you achieved previously or successfully model a more complicated relationship

## This is a group project
It's crucial to get some experience working in a team, so this project will be worked on in groups of >= 2.
You are expected to practice pair programming, communication, teaching each other and using Git & GitHub collaboratively to make branches and pull requests. We expect more than 75% of the code to be developed in pair programming sessions.

## Cloud training credits
You are expected to ask your instructor for access to cloud computing resources (EC2) to train your model.

## Deliverables
- GitHub repo with all code, model and README
    - it doesn't matter who is the owner of the repo, but you should each be collaborators with roughly equal input
- 10 minute presentation on your problem, approach & results
    - you should split the presentation between your team

## Milestones

### Milestone 1 (2 days after launch)
- Write a 1-2 paragraph memo explaining your idea and why you think it is worth working on. 

Don't go too far trying to develop a novel or a full theoretical specification. Here's an example of what we expect:

"""A huge problem for online fashion retailers is pricing products competitively. It can have a huge influence on their primary business metric: revenue. I propose that we build a deep learning system for pricing products, using the fashion clothing dataset collected by Jennifer in the first project. Making this kind of prediction can't be done by relying on simple features like the weight or material. A lot of information is contained within the product images. Hence the need for deep learning; we'll develop a CNN that predicts price from the images.

I personally want to work in the fashion industry as a data scientist/ML engineer, so I think that this project will be a great showcase of relevant skills. This problem is also a widely used practical application of AI, so I think it will me (and you!) develop transferrable skills relevant to many other roles.
"""

"""
I want to build a GAN for producing images of new trainers. I've found a [dataset](https://www.kaggle.com/sebastiaanjohn/sneakers) of loads of trainers that we can use. Generative design is an up and coming trend.

There are many different things that we could train a GAN to generate images of, but I chose to focus within the fashion space because retail companies have pretty mature AI systems for basic things (like product recommendations) and are best poised to start using this technology (some advanced companies like Thread or StitchFix already are!). Although generative design is not widely used across companies in their designing, building a GAN is a good showcase of many other skills which have wider application, such as using CNNs.

I also think that this project would produce a very visual result which could be deployed in a webapp. This would help us to get noticed more easily if we build a product like a recommendation system which might be harder for people to play with.
"""

## Marking criteria

Each of the following will be scored as "not attempted" (0 points), "attempted" (1 point) or "successfully applied" (2 points).

- PyTorch dataset and dataloader
- Neural network model
- Data augmentation
- Some hyperparameter tuning
- Monitoring loss curves (and visualising results if necessary) using Tensorboard
- GPU acceleration
- Evaluation of results

On top of this, there will also be points available for:
- initial dream press release (5 points) [added late]
- collaboration displayed and use of Git & GitHub (5 points)
    - branch protection implemented
    - use of GitHub projects with labels, milestones, assignees
- code cleanliness and structure (5 points)
- presentation of results (5 points)

The final score will be your percentage of the total available points attained.

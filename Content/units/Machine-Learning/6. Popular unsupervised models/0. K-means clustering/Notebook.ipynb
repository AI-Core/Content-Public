## Introduction: Unsupervised Clustering

>The K-means clustering algorithm can be used for supervised classification or unsupervised clustering.

Thus far, we have only discussed the supervised-learning case in which the dataset includes labels. Contrarily, in the unsupervised-learning case, the dataset does not include labels.
Here, the goal is not to predict a label, but to uncover the underlying structure of the data.


## Clustering
One approach for characterising the structure in data is to group different examples into clusters; this is known as **clustering**.

There are many ways to perform clustering, and each of them may produce different clusters. In this notebook, we explore the K-means clustering algorithm.

## The K-Means Clustering Algorithm
In the K-means clustering algorithm, K random points called centriods are initialised in a feature space and are each associated with a particular cluster. Subsequently, each example is assigned to be a member of the cluster whose centroid it is closest to. Afterwards, the position of each centroid is updated to be the average of all of the examples that are currently assigned to it.
These processes of updating the centroid positions and assigning examples continue until the assignments of examples remain constant and the algorithm converges.

![](images/kmeans.gif)

## K-Means Clustering Pseudocode
- Initialise K points called centroids in feature space, which represent the central point of a certain class.
- Until convergence,
    - assign each datapoint to the class represented by the nearest centroid.
    - update the centroid position to the average feature vector of all of the datapoints currently assigned to initialise the process.

## Limitations of the Algorithm
- If a centroid is not the closest to any datapoint, it will not be updated and will represent a random point in space.
    - This may not be noticeable if the data have an overly high dimension, hindering visualisation.
- In the unsupervised case, it may be difficult to select the best K to use.
    - However, it can also be an advantage in cases where the output clusters are to be used as inputs for further processing. For example, in image processing, K-means can be used to convert an image into monochrome before processing it.
- K-means does not always provide the same results because the centroids are placed randomly.
- As the number of dimensions of each example's features increases, the distance between any pair of datapoints tends to be constant; this reduces the effectiveness of the algorithm as it becomes increasingly difficult to distinguish examples. This is the curse of dimensionality at work. For more information, see [here](https://developers.google.com/machine-learning/clustering/algorithm/advantages-disadvantages).
- K-means clustering cannot cluster groups of data that are not blob shaped. This is because some of the K-nearest points may not be part of the same cluster. See the graph below, where K-means clustering fails to correctly cluster the datapoints.

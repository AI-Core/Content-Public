## What does the Central Limit Theorem tell us?

- It states that the distribution of the sample means will be normally distributed, regardless of the sample size, as long as the population is normal
- It states that the distribution of the sample means will be normally distributed, only if the sample size is large enough, regardless of the population distribution ***
- It states that the distribution of the sample means will always be skewed, regardless of the sample size or population distribution
- It states that the distribution of the sample means will be uniform, regardless of the sample size or population distribution

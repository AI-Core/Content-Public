## You are done applying, but you are not hired yet! 

Now it’s time to prepare for that all-important interview that stands between you and your new role!

Preparing for an interview can be a daunting task, but there are several simple things you can do to increase your chances of success.

### Research the company
Before the interview, take some time to research the company. Look at their website, read about their mission, and learn about their products or services. This will help you understand the company's culture and values and allow you to better tailor your answers to the interview questions.

### Review the job description
Study the job description carefully and make a list of the required skills and qualifications. This will help you prepare specific examples of how you meet these requirements and can excel in the role. So many people do not do this..

### Prepare for common questions
Review common interview questions and practice answering them out loud. This will help you feel more confident during the interview and ensure that you are prepared to answer questions that may arise.

### Practice with a friend
Practice your interview skills with a friend or family member. This will help you get comfortable with the process and give you the opportunity to receive feedback on your answers. And don’t forget to book your mock interview

### Dress appropriately
Even if the company has a casual dress code, it's important to make a good first impression. It doesn’t matter if it’s a remote interview, you should still dress appropriately. Avoid being too colourful and flashy.

>Pro tip: Benjamin Franklin said, “By failing to prepare, you are preparing to fail”

## Book your mock interview in the practicals

The purpose of a mock interview is to help you practice your interview skills, build confidence, and receive feedback on your responses. It is important because it allows you to identify areas where you need to improve, such as body language, communication skills, and answering interview questions.

During a mock interview, you can also practice answering tough questions that you might encounter in a real interview. You will learn how to handle difficult questions, such as questions about your weaknesses or employment gaps.

## Prepare your own questions for the interviewer

Asking questions during a job interview is an important way to demonstrate your interest in the position, learn more about the company and the job, and showcase your skills and experience. It can help you make a positive impression on the interviewer and increase your chances of landing the job.

Here’s a list of questions you can ask potential employers, use your own judgement as to which questions best fit. Your conversation should always be natural. There’s no particular order to ask these questions, and by no means should you ask all of them.

- Can you describe a typical project or product lifecycle at your company?
- Can you tell me about your company culture and how it supports the development team? 
- How do you measure success for your engineering teams?
- What makes your best employee stand out?
- What’s the vision of the company?
- Is there anything I can do to improve my skillset in the meantime?
- Is there anything about me that you are unsure about? If so, I would like to address any concerns.


## Job interviews can be stressful, and many people make mistakes during the process that can hurt their chances of getting the job. 

Here are the top 5 mistakes that people tend to make in interviews:

1. Not preparing enough: Many candidates don't do enough research on the company or the position, don't practice common interview questions, and fail to prepare examples of their skills and experience. It's important to thoroughly prepare for the interview.

2. Being late or unprepared: Showing up late or forgetting to bring important materials like your resume or work samples can make a bad impression on the interviewer. Be sure to arrive on time and bring everything you need.

3. Poor body language: Your body language can say a lot about you during an interview. Avoid slouching, fidgeting, or crossing your arms, as these behaviors can make you seem disinterested or unconfident.

4. Not answering questions fully: Listen carefully to the interviewer's questions and answer them fully. Avoid giving short answers or straying off topic. Use specific examples to illustrate your responses.

5. Speaking negatively about previous employers: Criticizing past employers, colleagues, or managers during the interview can reflect poorly on you. Instead, focus on the positive experiences and skills you've gained in previous roles.

By being aware of these common mistakes and taking steps to avoid them, you can improve your chances of acing your next job interview. Good luck!

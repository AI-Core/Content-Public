# How to beat the ATS (Applicant Tracking System)
Applicant Tracking Systems (ATS) have become an integral part of the hiring process, and their prevalence is not just a passing trend. Recent reports indicate that approximately 75% of companies, including almost all Fortune 500 companies, utilize ATS. Fortunately, navigating past the ATS bot and ensuring your resume reaches the hands of a hiring manager is not as daunting as it may seem. This comprehensive guide will equip you with everything you need to know about conquering Applicant Tracking Systems and maximizing your chances of securing an interview.

## What exactly are Applicant Tracking Systems?
Applicant Tracking Systems are automated software programs designed to scan resumes for specific keywords and eliminate those that do not meet the predefined criteria. However, it is important to note that these systems are not flawless. In fact, many hiring managers acknowledge that some qualified candidates may unintentionally be filtered out.

## How do Applicant Tracking Systems operate?
Let's take a closer look at how an Applicant Tracking System functions:
- A company has an open position for a Data Engineering role that requires Python experience.
- The job opening is posted on popular job search platforms like Indeed.
- The hiring manager receives over 250 applications from individuals with various backgrounds, including sales, marketing, human resources, software and data engineering.
- The company's ATS scans all the resumes for the keyword "Python." "SQL" and "Databricks".
- The ATS filters out the CVs which are badly formatted.
- The ATS also filters out CVs with images and symbols.
- The ATS presents resumes that contain the relevant keywords "Python" "SQL" "Databricks" and excludes those that do not.
- The hiring manager focuses solely on the resumes that mention Python.

The hiring manager is now able to review a manageable number of resumes, avoiding wasted time on applicants who lack the necessary experience. However, there is a potential downside. Some exceptional candidates may be inadvertently discarded because the ATS could not identify the desired qualifications (more on this later).
Despite the inherent risk, an increasing number of companies are adopting ATS. This is where our guide comes in - our objective is to minimize the number of qualified job seekers who are unfairly rejected.

## A step-by-step guide to outsmarting Applicant Tracking Systems
Although it may appear complex, effectively maneuvering past resume tracking systems is not as challenging as it seems. 

### Let's break it down into simple steps:
- Begin by creating your resume using Microsoft Word or Google Docs. While other text editors are acceptable, avoid using elaborate image-based resume programs.
- Utilize the provided standard resume template.
- Tailor your resume to each specific job application by incorporating relevant keywords from the job description.
- Thoroughly proofread your resume, meticulously checking for errors in spelling and grammar.
- Ensure that your resume is compatible with ATS programs by uploading it to a free ATS resume scanner.
- Limit the number of positions you apply for within the same company, focusing on roles that align with your qualifications.
- Always follow the application instructions, even if they seem contradictory to general best practices.

## The key to conquering ATS
Looking for a more straightforward approach? Here it is!

### Beating Applicant Tracking Systems boils down to two essential elements: readability and relevancy.
But what does that mean in practical terms?
- Readability: Ensure your resume is in an ATS-friendly format that enables resume tracking systems to comprehend the information you provide.
- Relevancy: Include pertinent details on your resume that demonstrate your suitability for the job, allowing the ATS to recognize your qualifications and pass your resume to the hiring manager.

Now, let's delve into these core principles with a bit more depth.

## Tips for crafting an ATS-beating resume
To ensure your resume is easily interpreted by Applicant Tracking Systems, simplicity is key. 

Here's what you need to do:
- Opt for a chronological resume format: ATS algorithms are designed to favor chronological resumes. List your work history in order, starting with your most recent position. Functional resume formats, which omit dates or job titles, can confuse ATS (and even human recruiters aren't fans of them!).
- Utilize a basic template: Don't fall into the trap of thinking you need a fancy or creative template to stand out. Keeping it simple is the best strategy.
- Stick to a single-column layout: While some ATS can handle multi-column layouts, they can still encounter glitches and misinterpretation. For example, if your work history is in one column and your education in another, the ATS might consider them a single section, failing to comprehend your experience accurately. Therefore, it's safer to structure your resume using tabs and left/right alignment, avoiding tables.
- Stick with standard resume fonts: Opt for a screen-friendly sans serif font like Arial, Calibri, or Garamond. Some newer fonts convert letters into special characters, which are often misinterpreted by ATS. Avoid using downloaded fonts, especially when submitting a Word document, as the hiring manager may not have the same font installed, causing your resume to appear garbled to both the resume scanner and the human recruiter. 
- Simplify your formatting: Don't stress too much about making your resume visually appealing. Not only do most recruiters prefer a straightforward layout, but fancy additions can also confuse Applicant Tracking Systems. 

### Specifically, avoid using:
- Images
- Tables
- Text boxes
- Graphics
- Logos
- Photographs
- Columns
- Graphs and charts
- Headers and footers for essential information (most ATS will skip these)
- Embedded hyperlinks (it's acceptable to include links in plain text instead)
### To make important information stand out, you can use:
- Bold
- Italics
- Underline
- Capital letters
- White space
- Section breaks
- Bullet points
- Color (note that different colors won't display in ATS programs, but they won't cause any issues)

### Export your resume in PDF format
It's generally recommended to submit your resume as a PDF. Why? Different versions of Word can alter your layout, fonts, and formatting, risking the integrity of your carefully crafted resume. PDFs are also easily shareable, as recruiters and hiring managers don't need to have Office installed to view your resume.

However, when specifically instructed to submit in Word, Google Docs, or another format, follow those instructions rather than saving as a PDF.

## What to include in your resume to beat ATS
Now that you know ATS can read your resume, it's crucial to ensure it captures the right information. Tailoring your resume to match the skills and experience the employer is seeking will increase its chances of scoring high enough to pass the ATS and reach a recruiter.

To get your resume past ATS, follow these guidelines:

### Use the appropriate keywords
The most effective way to beat ATS is by using the right keywords. How do you determine which keywords are the right ones?
Look at the job ad! Employers often provide a list of "must-haves," including specific skills and experience they seek.
Save the job description when you apply to have it as a reference, as the opening may be taken down by the time you reach the interview stage.

### Avoid keyword stuffing
Including keywords is good, but stuffing your resume with them is not. What's the difference? Use keywords in appropriate contexts and avoid excessive repetition. It's better to naturally incorporate a handful of keywords throughout your resume than to have an entire section filled with gibberish. While excessive keyword stuffing may help you pass an ATS, it will likely lead to your resume being discarded as soon as it reaches a recruiter's desk. Identify the most relevant or "high-value" keywords and use them sparingly, never sacrificing the readability of your resume for the sake of including extra information.

### Customize your resume
There are various ways to tailor your resume, some easier than others:
- Use a resume title that matches the job you're applying for.
- Clarify your job titles if they don't exactly match the desired role.
- Adjust your skills section to align with each job application.
- Customize your work experience section by highlighting relevant jobs and accomplishments.

### Employ standard section titles
Avoid using non-standard section titles on your resume. Stick to clear and straightforward headings like "Work Experience" that the ATS and hiring managers can easily recognize and navigate. Make it easy for them to understand your resume without any unnecessary confusion.

### Spell out abbreviations and acronyms
ATS systems are programmed to search for exact keyword matches. To increase your chances of getting past the ATS, spell out abbreviations and acronyms whenever possible. Include both the full form and the abbreviated version if the keyword is an acronym. This ensures that your resume matches the specific language used in the job description.

### Avoid buzzwords
ATS systems do not search for buzzwords or soft skills like "hard-working" or "flexible team player." Don't waste space on your resume by including these buzzwords. Focus on highlighting your relevant skills and experience instead. Remember, recruiters are looking for substance, not clichés.

### Check your spelling and grammar
Always double-check your resume for spelling and grammar errors. While a human reader might be able to decipher a typo, an ATS will not. Submitting a resume with errors can lead to rejection, so take the time to ensure your document is error-free. It's a simple step that can make a big difference in your application's success.


## Tips for career changers: Utilize a resume summary
If you're transitioning to a new career and lack direct experience in the desired role, incorporating a resume summary can be beneficial. This concise paragraph allows you to highlight your relevant skills, transferable abilities, and aspirations.
The resume summary provides an opportunity to include keywords and skills that may not fit elsewhere in your resume. Rather than falsely claiming experience in a similar role, emphasize your transferable skills. Additionally, if you've acquired new skills or completed relevant qualifications, emphasize them in the summary and include them in the education section.

## Common ATS errors (and how to avoid them)
There are two main reasons why your resume might be rejected by an ATS:
- Insufficient background or experience matching the hiring manager's requirements
- An error or mistake in the application.
While little can be done about the first reason, the second can be easily resolved. 

### Here are some common reasons an ATS might mistakenly reject your resume and tips to avoid such issues:
- Error:
The ATS cannot read your resume.
- Solution:
Refer to the formatting tips mentioned earlier in this guide.
Remove any non-standard formatting, downloaded fonts, or images.
Follow the steps below to test your resume against ATS scanners.

- Error:
The ATS did not identify an essential keyword.
- Solution:
Review the job description to identify must-have skills.
Utilize skills and keyword finders or Targeted Resume tools to suggest relevant keywords.
Ensure proper spelling of all keywords.

- Error:
The ATS perceives a lack of relevant experience.
- Solution:
Incorporate the exact job title somewhere on your resume.
Clarify previous job titles if they don't align with the new role.
Spell out acronyms and use wording consistent with the job description.

## Evaluate if your resume is ATS-proof
Let's address a common misconception: There is no foolproof method to guarantee that your resume will pass an ATS. This is due to the numerous factors involved, such as the wide range of ATS software and the specific criteria they're programmed to evaluate.

### However, you can perform a few simple tests to determine if your resume can overcome common ATS obstacles. Here's how to assess your resume:
- Create your resume in Microsoft Word, Google Docs, or a similar word processing program.
- Avoid using image-based resumes created with Canva or Photoshop, as most ATS cannot read them.
- Do not scan and upload your resume, as it can make it challenging for the ATS to read.
- Open your resume in Chrome, Preview, or Adobe Acrobat Reader and try to highlight/select the text. This mimics how ATS "read" resumes. If you cannot highlight the text, neither can the ATS.

## Do's and don'ts of bypassing ATS (aka how to beat the system)
Looking for a guaranteed method to outsmart the ATS? Well, there isn't one.

- DON’T attempt to deceive or trick the ATS. Despite some suggestions, there is no way to "cheat" Applicant Tracking Systems.
- DON'T copy and paste the entire job description in small, white text at the bottom of your resume. ATS will remove the formatting, making it evident that you attempted to manipulate the system.
- DON'T create a separate "keywords section" or repeatedly copy and paste various versions of the same keyword. While you might pass the initial resume tracking phase, you're likely to be swiftly rejected by the hiring manager.

- DO follow the previously mentioned steps to ensure that Applicant Tracking Systems can effectively read your resume.
- DO invest time in tailoring your resume to impress not only the ATS but also the hiring manager.

## FAQ

### How prevalent is the use of ATS among companies?
According to various surveys conducted in recent years, the overall adoption of Applicant Tracking Systems (ATS) by companies falls somewhere between 55% to 75%. What we can be certain about is that a majority of companies employ ATS, with larger organizations being more likely to utilize them. In fact, an astounding 99% of Fortune 500 companies now rely on Applicant Tracking Systems.

### How many different types of Applicant Tracking Systems exist?
Determining an exact number is challenging, but current estimates suggest that there are approximately 350 distinct resume tracking systems available.

### Which are the most common Applicant Tracking Systems?
Some of the popular ATS software commonly used include Workday, LinkedIn Recruiter, Recruiterbox, Jobvite, iCIMS, Taleo, and LeverTRM.

### Are there differences among Applicant Tracking Systems?
While various types of ATS exist, their core purpose remains the same. You need not worry about identifying a specific ATS you may encounter. By following the aforementioned steps, you can ensure that your resume is readable by any ATS, regardless of the software used.

### What criteria do ATS focus on?
It's crucial to recognize that resume scanners are not entirely automated; there is always a human element involved in the ATS software. The hiring manager determines the non-negotiable skills or experience they are seeking and configures the ATS to scan for those specific keywords. ATS typically scan for hard skills, meaning keywords like "Azure" or "Pytorch" might be relevant, whereas soft skills like "time management" may not be considered.

### How do ATS algorithms operate?
The functioning of ATS algorithms depends on how each company has chosen to implement their ATS. Most ATS employ basic filters, such as scanning for resumes containing the keyword "sales" or identifying resume gaps by examining employment dates. Sometimes, resumes that do not match the ATS algorithm may be automatically discarded, while in other cases, these keywords are flagged to facilitate hiring managers' quick review or the resume may receive a preliminary score to expedite the recruitment process.
ATS can also be utilized early in the job application process. When submitting an online application, you often encounter questions about your background and availability. Some of these questions act as knockout factors. For instance, if an application asks whether you possess a particular qualification or are willing to travel, responding negatively to these questions can lead to automatic rejection.

### How effective are ATS?
According to a report from Harvard Business School, over 80% of employers acknowledge that qualified candidates are being excluded from the hiring process because their skills and experience do not exactly match the criteria within the applicant tracking software. This highlights the importance of using appropriate keywords. If in doubt, always refer to the job advertisement, as ATS keywords are likely to utilize similar phrasing.

### Why do companies opt for ATS?
Despite its flaws, the primary reason companies use ATS is to save time and resources. On average, a job posting receives over 250 applications, and many are the result of candidates applying en masse to semi-relevant positions. Instead of allocating an employee to manually review each application, an ATS can swiftly scan resumes and filter out unqualified applicants, thereby narrowing down the pool to a more manageable size.

### Will a human review my resume?
Yes! The ATS serves as the initial step in the process. Once your resume successfully passes the ATS, it will be manually reviewed by a human recruiter or hiring manager. Consequently, there is no room for keyword stuffing or deceptive tactics, such as including a copy of the job description within your resume. The ATS does not decide who gets the job but rather determines who advances to the next stage of the selection process.

### Do ATS check cover letters?
Not usually, no. Cover letters are a more personal touch and a way for the hiring manager to get to know more about you, so don't worry about optimizing them for ATS.

### Can I cheat the system?
Nope, sorry! Beating Applicant Tracking Systems isn't about finding a shortcut — it's about including the right keywords and making sure ATS can easily read your resume.

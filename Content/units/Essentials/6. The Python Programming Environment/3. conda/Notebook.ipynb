# Conda


> Conda is an open-source environment and package management system. An environment in this context refers to a set of specific versions of libraries and tools that are isolated from others. With an environment manager like Conda, you can set up different versions of the libraries and tools you need for a specific project, allowing you to keep each project self-contained, and avoiding clashes between the requirements of different projects.

### Advantages of Conda

Conda offers several advantages as a package and environment manager:

- **Cross-Platform**: Conda is platform-agnostic, so you can use it across Linux, macOS, and Windows. This makes sharing and collaborating on projects much easier.
   
- **Multi-Language Support**: While Conda was created for managing Python environments, it can manage libraries in any language, making it versatile for multi-language projects
   
- **Ease of Use**: Conda's command-line interface is intuitive and user-friendly, making environment and package management straightforward

## Conda Environments

> Conda environments are isolated spaces where software packages and dependencies can reside, ensuring that different projects can have their own specific versions of libraries and tools without interference. 

Under the hood, a Python Conda environment is essentially just a directory that contains a specific collection of Python packages. By switching between environments, you can quickly and easily change the tools and libraries available to your projects. This is especially useful when different projects have different requirements, preventing potential conflicts between package versions.

### Why Use Conda Environments?

Using Conda environments offers several benefits:

- **Isolation**: Each environment is separate, so changes in one environment won't affect others. This is important when working on multiple projects with different requirements.

- **Reproducibility**: By exporting an environment's specifications, you can recreate it on another machine, ensuring consistent behavior across platforms and systems

- **Version Control**: Environments allow you to work with specific versions of packages and libraries, making it easier to manage and update them as needed

- **Flexibility**: You can easily switch between different environments, test new packages without affecting your main environment, and ensure that your projects remain unaffected by changes in dependencies

### Creating Environments

To create a new Conda environment, use the `conda create` command. For example, to create an environment named `myenv` with Python 3.11, you would use:

```bash
conda create --name myenv python=3.11
```

This command creates a new environment called `myenv` with Python version 3.11 installed. Conda will develop a plan for the environment and then ask you to confirm that you wish to go ahead by typing `y`:

### Activating and Deactivating Environments

Once you've created an environment, you need to activate it before you can use it. To activate the environment you have just created, use the following command:


### Installing Packages

> Conda acts as a Python *package manager* as well as an environment manager. A Python **package manager** is a tool that automates the process of installing, updating, and removing Python libraries. When you install a package using Conda, it does more than just download the package. Conda checks for dependencies the package might have and ensures that all required libraries are installed simultaneously. This dependency resolution helps prevent dependency conflicts, where two libraries each need different versions of a particular dependency.

To install a package in the active Conda environment, you use the `conda install` command. For example, to install the Python library `NumPy`:

```bash
conda install numpy
```

This command installs the latest version of `NumPy` available in the Conda repositories. When you run the command, Conda will decide which packages it needs to install to fulfil your request, and prompt you for confirmation, before downloading and installing the packages. 

<p align="center">
    <img src="images/install_numpy.gif"  width="800"/>

You can also choose to install a specific version of a package by adding the specific version number, or a set of version constraints:

```bash
conda install numpy=23.9.0
conda install  torchvision">=0.3.0"
conda install  openpyxl">=2.4.10,<=2.6.0"

```
It is also possible to bypass the confirmation step by including a `-y` flag to auto-approve the installation plan:

```bash
conda install -y numpy
```


### Installing Packages During Environment Creation

It is also possible to add packages during the environment creation step. Consider an imaginary scenario where you are about to undertake a data science project. You know you'll be wanting to use `pandas` and `NumPy` packages for data handling, and you also want to build some simple machine learning models using `scikit-learn`. Additionally, you wish to use `git`for version control. 

Instead of creating an environment and then installing each package one-by-one, you have a number of options for creating the whole environment in one go. 

One approach is to specify the packages you require as arguments to the `conda create` command:

```bash
conda create -n datascience_env git pandas numpy scikit-learn
```

Alternatively you can generate a `requirements.txt` file as follows:

````python
# requirements.txt
pandas
numpy
scikit-learn
git
````

Then you can pass this file as an argument to the `conda create` command by using the `--file` flag:

```bash
conda create -n datascience_env --file requirements.txt
```

## Advanced Conda Features

Conda is a powerful tool with many advanced features that can help streamline your workflow and manage packages and environments more effectively. 

### Conda Channels

> *Conda channels* are repositories where packages are stored and made available for users to install. When you request a package to be installed using Conda, it searches these channels to find the software you've requested. 

Most of the time you will only need to use the main Conda channel, but some packages that are not available on the default channel can be found on the `conda forge` channel, which is a community-maintained channel that often adopts new and lesser-known packages more rapidly than the default channel. 

- **Using Channels**: To install a package from a specific channel, you can specify the channel using the `--channel` (or `-c` for short) flag:

```bash
conda install --channel conda-forge some_package
```

- **Adding Permanent Channels**: If you find yourself frequently using a specific channel, you can add it to your Conda configuration so that it's always checked:

```bash
conda config --add channels conda-forge
```
While channels offer a broader range of packages, it's important to exercise caution in using them. Not all packages in non-default channels undergo the same level of testing or scrutiny. Always ensure you're getting packages from reputable channels to maintain the integrity and security of your environments. You are unlikely to need more than the default and `conda forge` channels for most purposes, although some specialist fields maintain their own community channels.

### Managing Conda Configuration

*Conda configuration* refers to the settings and parameters that control how Conda operates on your system. These configuration settings are stored in a configuration file, typically named `condarc` or `.condarc`, and are used to customize various aspects of Conda's behavior.

You can view your Conda configuration using:

```bash

```

### Cloning and Sharing Environments

It can sometimes be useful to create copies of an environment, either to test a new feature without affecting the primary environment, or so that you can share your environment with other users.

#### Cloning Environments

Cloning an environment creates an identical local copy of the environment on your machine. To clone an environment, you can use the following syntax:

```bash
conda create --name new_env_name --clone old_env_name
```



### Sharing a Conda Environment with Other Users

> Sharing a Conda environment is useful when collaborating on projects, as it ensures consistency and reproducibility across different setups. By sharing an environment, you can be certain that everyone is using the same versions of packages, which can help prevent "it works on *my* machine" issues.

#### Exporting the Environment

1. **Export to a `.yml` File**: The most common way to share an environment is by exporting it to a `.yml` (`YAML`) file. A `YAML` file is a human-readable format used to store structured data. In the context of Conda, it's often used to share environment configurations, allowing you to export a list of packages and their versions to a `YAML` file, which can be easily recreated on another system. 

    ```bash
    conda env export --name your_environment_name > environment.yml
    ```

    Replace `your_environment_name` with the name of the environment you want to share. This command will create a file named `environment.yml`.

2. **Share the `.yml` File**: Once you have the `.yml` file, you can share it via distributed version control systems like GitHub, or any other file-sharing method.

#### Importing the Environment

For the recipient to use the shared environment:

1. **Download the `.yml` File**: Ensure you have the `environment.yml` file on your machine

2. **Create the Environment from the `.yml` File**: Use the following command to create a Conda environment based on the `.yml` file:

    ```bash
    conda env create -f environment.yml
    ```

    This will create a new Conda environment with all the packages and versions specified in the `environment.yml` file.

3. **Activate the New Environment**: Once the environment is created, activate it using:

    ```bash
    conda activate your_environment_name
    ```

    Replace `your_environment_name` with the name specified in the `.yml` file or provide a custom name if you've chosen to during the creation process.

#### Tips for Sharing Environments

- **Keep It Clean**: Before exporting, ensure the environment contains only the necessary packages to reduce the size and complexity of the `.yml` file

- **Include a README**: When sharing an environment, it's helpful to include a `README` file with instructions on how to import and use the environment, especially for those unfamiliar with Conda

- **Version Control**: It is good practice to use version control systems like git to track changes to the environment over time. This can be especially useful for collaborative projects.

By sharing Conda environments, you ensure a consistent development environment across different machines, making collaboration smoother and more efficient.

## Conda Best Practices

To maximize the benefits of Conda and ensure a seamless experience, it's crucial to adhere to certain best practices. These guidelines will assist you in maintaining organized, efficient, and conflict-free environments.

### 1. One Project, One Environment

Creating a new environment for each project is a fundamental best practice. This approach ensures that dependencies for one project don't interfere with another, reducing the risk of version conflicts. Having distinct environments for each project simplifies the management and understanding of a project's dependencies.

### 2. Minimize Packages in an Environment

When setting up an environment, it's advisable to install only the packages necessary for your project. This strategy reduces potential conflicts and ensures the environment remains lightweight. It's better to resist the temptation to have a single environment with every conceivable package, as this can lead to version conflicts and make the environment challenging to manage.

### 3. Regularly Update Conda and Packages

Keeping Conda updated ensures you benefit from the latest bug fixes, performance improvements, and new features. Use the command `conda update conda` to update Conda itself. Additionally, it's a good habit to periodically update the packages in your environments to access the latest features, optimizations, and security patches. Be aware though that the decision to update packages in your environment in this manner may cause your code to stop working if the latest version of a library is not back-compatible with earlier versions. Updating the packages in your environment should be undertaken as part of a wider code reviewing process, refactoring any sections of your code that would be deprecated by the update.

### 4. Export and Version Control Environments

Regularly exporting your environments to `.yml` files is a valuable practice. These exports provide a snapshot of your environment, which can be instrumental for debugging, sharing, or transitioning to another machine. Additionally, storing your exported `.yml` files in version control systems like git allows you to track environmental changes over time and facilitates collaboration.

### 6. Clean Up Unused Environments and Packages

Over time, you might accumulate environments and packages that are no longer in use. Consider removing old environments to free up space using the command `conda env remove --name old_environment_name`. To remove unused packages and reclaim space, the `conda clean --all` command is beneficial.

### 7. Avoid Clashing Package Managers

When working with Python and other languages, it's common to encounter multiple package managers. For Python, besides Conda, there's `pip`, which is the default package installer. While both are powerful tools, using them simultaneously without caution can lead to conflicts and inconsistencies in your environments. You will learn about `pip` in another lesson, but for now the important message is to be aware of which manager installed a particular package. 

Mixing Conda and `pip` indiscriminately can result in overlapping packages, version mismatches, or even broken environments. Here are some guidelines to avoid these issues:

- **Prioritize Conda**: Whenever possible, install packages using Conda. Conda's environment management and dependency resolution are robust and can help prevent many common issues.

- **Use `pip` Carefully**: If a package isn't available through Conda or you need a specific version only available through `pip`, activate your Conda environment and then use `pip` to install the package. This way, the package will be contained within the Conda environment.

- **Document Non-Conda Packages**: If you do install packages with `pip` within a Conda environment, make a note of them. This documentation will be crucial if you need to recreate the environment or share it with others, as the `conda env export` command won't capture `pip`-installed packages by default. A `requirements.txt` file is a good place to store this information.

By consistently following these best practices, you'll ensure a more streamlined, organized, and hassle-free experience with Conda, allowing you to concentrate on your projects rather than on managing dependencies.

## What will be the output of the following Python code?

```python
x = 5
y = 10
if x > 3 and y < 15:
    if x + y < 20:
        print("x + y is less than 20")
    elif x + y == 20:
        print("x + y is equal to 20")
    else:
        print("x + y is greater than 20")
else:
    print("The condition is not met")
```

- `x + y is less than 20` ***
- `x + y is equal to 20`
- `x + y is greater than 20`
- `The condition is not met`

- The `elif` statements are executed first in the order they are written, followed by the `if` statement, and the `else` statement is executed last if none of the conditions are met

## What is the purpose of the break statement in a while loop?

- It ends the current iteration and continues with the next iteration
- It exits the while loop completely, regardless of the loop condition ***
- It skips the current iteration and continues with the next iteration
- It restarts the while loop from the beginning

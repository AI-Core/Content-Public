# Reports and Visualisations

> The main use case of Power BI is the generation of professional-grade reports and visualisations, so that analytics insights can be shared with decision makers, helping them to make data-driven decisions.

Whereas in the previous lessons have focused exclusively on Power BI Desktop, this lesson will relate additionally to the Power BI Service, which is the main software-as-a-service (SaaS) component of the Power BI ecosystem. Users can create, view and save reports for free in Power BI Desktop, but in order to share them with other users in Power BI service, both the creator and the recipient must have a paid license.



## Reports vs. Dashboards


### Reports 

> A *report* in Power BI is a collection of charts, tables, and other visualisations from a single data set. The components of a report are interactive and interconnected, meaning that selecting an item in one visual can affect the others. 

Reports allow users to explore their data and discover insights. Reports can have multiple pages, and you can drill down into the data for more detail. You can also use Power BI Desktop to create and publish reports to the Power BI service. Power BI reports offer a lot of flexibility in terms of data exploration and visualisation because they allow users to pivot the data in real-time and create different views.

- Typically generated in Power BI Desktop
- Reports refer to a single data set
- Can have multiple pages
- Can be made to be highly interactive
- Each page will have one or more visualisation *tiles* - graphical blocks containing a chart, table or other element


### Dashboards

> A *dashboard* is a consolidated view of visuals, reports, and other data pinned from different reports. A dashboard is contained on one page and is designed to present a high-level overview of business metrics and key performance indicators (KPIs).

Unlike reports, dashboards are not interactive in the same way. While you can interact with individual tiles, selecting something on one tile does not affect the other tiles. However, clicking on a tile in a dashboard can lead you to the underlying report or dataset for more detail.

- Created within Power BI Service
- Can contain visualisations extracted from multiple different reports
- Not interactive to the same degree as reports



## The **Report View**

> The **Report View** is where you do most of the work in designing and building reports. It contains all of the tools required to create visualisations and arrange them into reports, as well as applying filters.

The **Report View** is composed of five main areas:

1. **Canvas**: This is the main work area where you build your report by adding and arranging visuals.

2. **Data Pane**: This is on the right-hand side of the window. It shows all the tables, columns, and measures in your dataset. You use these to build visuals, either by dragging and dropping them onto the canvas or by selecting them and choosing the type of visual from the Visualisations pane.

3. **Visualisations Pane**: This pane is also on the right side and it contains all the visualisation tools you can use to represent your data. This includes bar charts, column charts, line charts, maps, and more. You select a visual type by clicking on the corresponding icon.

4. **Filters Pane**: Here, you can apply different filters to your data at various levels such as visual, page, and report level.

5. **Pages Pane**: Located at the bottom of the screen, this pane allows you to add, select, duplicate, or remove report pages. Each page can contain a different set of visuals.

You can interact with the data in the **Report View** by drilling down, applying slicers and filters, or changing the visual type. The overall goal in the **Report View** is to design an interactive data story that allows your audience to explore and understand the data.


## Elements of a Report

### Visualisations

> *Visualisations* in Power BI are graphical representations of your data. They include graphs like bar graphs or scatter plots, but also tables, and *cards*, a type of visualisation that gives prominence to an important metric or KPI. Visualisations allow users to visually explore and analyze data to derive meaningful insights. 



### Tiles

> A *tile* is a single visual, text box, image, or other element that you add to your report. It is the basic unit of a report and represents a specific piece of content. Tiles can contain visuals, text, images, tables etc, and are interactive.

- To add a new tile, double-click on the canvas
- You can resize tiles by dragging on the corners or vertices
- With a tile selected, you can determine the contents using the Visualisations pane

### Containers

> *Containers* in Power BI are elements that you can use to organise and manage other elements like tiles in a report. When items such as visuals (tiles) are grouped together, they move and resize together, maintaining their layout and relative position. This can be extremely useful when designing complex reports or dashboards.


### Filters

>*Filters* are used by report designers to restrict the  subset of data that is displayed in section of a report, via the application of logical rules. They remove unwanted parts of the data set to help users focus on the information that is relevant to their analysis. There are three different granularity levels of filter: report, page, and visual. They are all accessed in the same way, via the filter section on the right hand pane.

Filters are not intended to be altered by consumers of the reports, and hence are not interactive in the completed report.


### Slicers

> *Slicers* are a type of filter but presented in a more user-friendly way, intended for the use by the report consumer. They are visual elements that are embedded in reports or dashboards, allowing end-users to filter data in an intuitive manner. Unlike filters which are typically used during the report design phase, slicers are typically used by end-users during the consumption of the report or dashboard.





## Types of Visualisation

### 1. Charts

> Charts in Power BI are visual representations of data that help to reveal trends, patterns, and insights that are often difficult to discern from raw data. These include a wide variety of chart types, such as bar graphs, line graphs and scatter plots, as well as more complex visualisation types. Power BI's interactive charts allow users to drill down into data for more detailed insights, highlight specific data points, and even combine multiple chart types into a single, interactive visual. As well as the range of charts that are available by default, additional visualisations are available to download via the [Microsoft Marketplace](https://appsource.microsoft.com/en-us/marketplace/apps?product=power-bi-visuals&exp=kyyw).

### 2. Cards

>A *card* is a type of visualisation that displays the single aggregated value (eg. a sum, count, average etc) of a specific field. Cards are useful for highlighting key metrics or KPIs because they allow users to see these important numbers at a glance. They are often used in dashboards or reports to provide a clear and concise summary of performance or results.

#### Sections of a Card 

- **Data label**: The main element of the card and it represents the aggregated value of the field you've chosen
- **Category label**: Displays the name of the field or measure that's being displayed
- **Conditional formatting**: You can apply conditional formatting to cards in Power BI, e.g. changing the colour of the data label based on its value

*Multi-row cards* are a variation of the card visual that allow you to display multiple data points in a tabular format. Each row in the table displays a field value, and you can include multiple rows in the same card.

### 3. Tables

> In Power BI, a *table* is a type of visualisation that displays data in a classic tabular format, organised into rows and columns. Each column in the table corresponds to a field from your dataset, and each row corresponds to a record. Ultimately, despite containing rows and columns, tables are a 1-dimensional format in Power BI, as all of the left-hand columns merely segment the value of the rightmost, aggregated column.

### 4. Matrices

> A *matrix* is a type of visualisation that is similar to a table, in that it displays data in a grid format. However, unlike a standard table, a matrix allows you to have hierarchies and display data across both rows and columns. 

Matrices allow us to solve the problem described in the section on tables - we can now view at a glance the total units sold for each segmentation category (in this case month and product category), independently of the other.

We have already covered the concept of a slicer earlier in the lesson. Slicers are design elements that live in tiles like any other visualisation. This contrasts with ordinary filters, which have their own dedicated pane and are not meant to be interacted with by people viewing the report.

### 6. Q&A visuals

> A *Q&A visual* is an interactive feature that allows users to ask questions about their data in natural language and receive answers in the form of relevant data visualisations. It's essentially a tool that uses a machine learning model to interpret the user's question, query the underlying dataset, and present the answer as a chart, graph, map, or other visual.

### 7. Smart Narratives

> A *Smart Narrative* is a visualisation that provides automatically generated, customisable text insights about your data. Using AI, the Smart Narrative feature analyzes your data and creates a summary of key points, trends, or interesting patterns, which is displayed as a dynamic text summary within your report.

For example, if you have a visual showing sales data over time, a Smart Narrative might summarise the overall trend ("Sales have increased 10% year over year"), call out the month with the highest sales ("The highest sales were in December"), or identify any unusual data points ("Sales in June were unusually low").

The Smart Narrative updates automatically as your data changes or as you interact with other visuals in the report. It's a powerful tool for making your reports more informative and engaging, helping users understand the key takeaways at a glance without having to interpret the visuals themselves.

## Creating an Example Report

In the lesson on the Power Query Editor we created a Power BI project called `Power_BI_demo_session.pbix`. We subsequently designed the data model for this project in the lesson on Reports and Visualisations.  

In this lesson we will use the same project to demonstrate how to create an example report. Open this project inside Power BI Desktop to follow along with the next sections of the lesson. If you have not completed the exercises in the previous lessons, please do so before continuing.


### Create a Graph of Orders per Month

First, let's make a graph showing how many orders occurred in each month.


- Open the **Report View**

<p align="center">
<img src="images/report_view_icon.png"  width="100" > 
</p>
<br><br>

- Double-click in the canvas to create a new tile
- Select **Scatter chart** from the `Visualisations` pane
- Drag `dim_datetime[month]` to the x-axis field
- Drag  `CustomerID`from the `Sales` table into the y-axis, and make sure the summary is set to **Count of** using the dropdown menu.

You should now see the chart displayed in the tile in the canvas. However the labels don't look quite right. Currently the title and y-axis legends refer to `count of CustomerID`, when in fact we want it to describe the number of orders per month.



- Open the **Format** pane
- Choose the **Title** dropdown, and edit the title to say `Orders per Month`

### Create a Graph of Units Sold per Product 

Now let's add a bar chart of the number of units sold in each product category. 

- Double-click the canvas again to create another new tile
- Drag the `dim_product[Product]` into the **X-axis** field
- Drag the `Sales[units]` field into the **Y-Axis** field
- Right click the `Sum of Units` value in the **Y-Axis** field and choose **Rename for this visual** from the menu
- Rename the value to `Units Sold`

### Create a matrix of units sold per segment and per month

We now want to add a table that shows a breakdown of how the number of units sold by product segment and by month. As we have two variables we want to compare, the **Matrix** visualisation is the best choice.

- Create another new tile, and select the **Matrix** visualisation
- Drag  `dim_datetime[Month]`into the **Rows** field and `dim_product[Segment]` into the **Columns** field
- Drag `Sales[Units]` into the `Values` field, and ensure the aggregation is set to **Sum of** using the dropdown.

Finally, let's add a slicer so that those who are looking at the report can see how the various breakdowns we have charted vary across days of the week. 

- Create another new tile and select the **Slicer** visualisation type
- Drag `dim_datetime[Day of Week]` into the **Field** field

## Publishing the Report

Once you have finished creating a report in Power BI Desktop, you can publish it to any Power BI service workspace that you have access to:

- Before you can publish a report, you need to save it. Go to **File** > **Save As** and give your report a name.
- Click on **File** in the menu bar, then select **Publish** > **Publish to Power BI**
- You will be prompted to sign in to your Power BI account. If you're not already signed in, you will need to enter your Power BI account credentials
- You will then be asked to choose where you want to publish your report. You can choose to publish it in `My Workspace` or in other workspaces where you have the appropriate access rights

## Key Takeaways

- Reports are interactive, multi-page visualisations of a single dataset, allowing users to explore data and derive insights
- Dashboards are consolidated, less-interactive views of visuals and data from multiple reports, designed for a high-level overview of metrics and KPIs
- The **Report View** in Power BI includes tools for creating visualisations, arranging them into reports, and applying filters
- Filters restrict the data displayed in a report section, providing focus on relevant information
- Slicers are interactive visual elements that allow end-users to filter data intuitively while viewing the report
- There are many types of Power BI visualisation including charts, cards, slicers, Q&A Visuals and Smart Narratives
- Once created, reports can be shared to a workspace in Power BI Service

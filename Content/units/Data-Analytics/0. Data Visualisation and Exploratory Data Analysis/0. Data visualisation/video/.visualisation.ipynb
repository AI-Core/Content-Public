# Data Visualisation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
df = pd.read_csv('.Life_Expectancy.csv')
```

```python
df_japan = df[df['Country'] == 'Japan']
sns.scatterplot(x='Year', y='Life expectancy ', data=df_japan)
plt.title('Life Expectancy by Year')
plt.show()
```

```python
df.groupby('Year')['Life expectancy '].mean()
```

```python
life_by_year = df.groupby('Year')['Life expectancy '].mean()
years = sorted(df.Year.unique())
sns.scatterplot(x=years, y=life_by_year)
```

```python
sns.histplot(x=df['Alcohol'], bins=30)
```

```python
sns.histplot(x=df[df['Country']=='Afghanistan']['Alcohol'], bins=10)
```

```python
sns.histplot(x=df[df['Country']=='Spain']['Alcohol'], bins=10)
```

### Heatmap

```python
numeric_columns = df.select_dtypes(include=np.number).columns
df_num = df[numeric_columns]
df_num = df_num.drop(['Income composition of resources', ' thinness  1-19 years', ' thinness 5-9 years'], axis=1)
corr = df_num.corr()
sns.heatmap(corr, cmap='coolwarm')
```

## Anscombe's Quartet

```python
import seaborn as sns
df_anscombe = sns.load_dataset("anscombe")
round(df_anscombe.groupby('dataset')['y'].std(), 2)
```

```python
print('Mean of X and Y')
display(round(df_anscombe.groupby('dataset')[['x', 'y']].mean(), 2))
print('Standard deviation of X and Y')
display(round(df_anscombe.groupby('dataset')[['x', 'y']].std(), 2))
```

```python
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df_anscombe,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

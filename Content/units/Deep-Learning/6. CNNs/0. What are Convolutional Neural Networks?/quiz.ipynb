# What Are Convolutional Neural Networks?

## True or False: Traditional “vanilla” networks can process tensors.

- True 
- False ***

## Convolutional NNs are useful because:

- They preserve the spatial relationships between elements in their inputs ***
- They require less memory than traditional neural networks
- They are faster to train
- They can handle larger datasets

## The output of a CNN is a:
 
- Vector
- Tensor ***
- Scalar

## If your input data are vectors of unrelated numbers, it is better to use a:

- "Vanilla" or feedforward Neural Network ***
- Recurrent Neural Network
- Convolutional Neural Network

## True or False: Parameter sharing is the process of using the same weights over different locations in an image.

- True ***
- False

## The depth (3rd dimension) of an output tensor generated by a 2D convolutional network represents:

 - The width of the input image
- The number of colour channels in the input image
- The number of filter kernels in the convolutional layer ***
- There is no depth to the output tensor, it is a 2D matrix

## What is the primary purpose of using convolutional layers in a Convolutional Neural Network (CNN)?

-  To reduce the spatial dimensions of the input data
- To detect local features and patterns in the input data, such as edges and textures ***
- To perform non-linear transformations on the input data
- To increase the depth of the network

 # AWS ML Services Overview

## Introduction
 AWS has a multitude of prebuilt AI services, which are prepackaged to be integrated into any organisation's workflow. Each service is custom-built for a specific task and easy to use. Further, in some cases, only the required data and no prior machine-learning knowledge are required for implementation. The use cases of these services range from creating image recognition applications, text to speech and text extraction to a fully comprehensive machine-learning suite Amazon SageMaker. The figure below provides information on many of the machine-learning services offered by AWS, with descriptions and use cases. To explore additional services, visit the [AWS management console](https://aws.amazon.com/console/).

```python
import requests
response = requests.get('https://www.currys.co.uk/gbuk/computing-accessories/components-upgrades/graphics-cards/324_3091_30343_xx_xx/xx-criteria.html')
response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text)
print(soup)
```

<p align=center><img src=.images/curry_1.png width=850></p>

```python
container = soup.find_all(name='div', attrs={'class': 'ProductListItem__DivWrapper-sc-pb4x98-7 cgxObq'})
len(container)
```

<p align=center><img src=.images/curry_2.png width=850></p>

```python
price_list = []
for i in container:
    price_element = i.find(name='span', attrs={"class": "ProductPriceBlock__Price-eXioPm eTWvaA"})
    price = price_element.text
    price_list.append(price)
print(price_list)
```

```python
from selenium import webdriver

driver = webdriver.Chrome()
```

```python
driver.get('https://www.currys.co.uk/gbuk/computing-accessories/components-upgrades/graphics-cards/324_3091_30343_xx_xx/xx-criteria.html')
```

```python
cookies_button = driver.find_element_by_xpath('//button[@id="onetrust-accept-btn-handler"]')
cookies_button.click()
```

```python
container = driver.find_elements_by_xpath('//div[@class="ProductListItem__DivWrapper-sc-pb4x98-7 cgxObq"]')
price_list = []
for i in container:
    try:
        price_element = i.find_element_by_xpath('.//span[@class="ProductPriceBlock__Price-eXioPm eTWvaA"]')
        price = price_element.text
        price_list.append(price)
    except:
        price_element = i.find_element_by_xpath('.//span[@class="ProductPriceBlock__Price-eXioPm ihovpx"]')
        price = price_element.text
        price_list.append(price)
print(price_list)
```

```python
next_button = driver.find_element_by_xpath('//a[@title="next"]')
next_button.click()
```

```python
search_bar = driver.find_element_by_xpath('.//input[@class="Input__InputElement-kaiOny fQGpYl"]')
search_bar.click()
```

```python
search_bar.send_keys("SD card")
```

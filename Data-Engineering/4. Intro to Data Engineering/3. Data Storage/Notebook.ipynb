# Data Storage

## What is Data Storage?

> Computer data storage is any recording medium that's used to retain digital data. Storage fundamental component of computing and data pipelines.

In the digital world, everything we interact with and work on, creates data. All the files, images, audio/video data that we daily watch and use generates an increasing quantity of data. This data needs to be stored somewhere to be able to access and analyze it later. As some experts say, data has become as valuable as gold in the modern age.

Storing data is very crucial and has led to development of many technologies. During late 1970’s and beginning of 1980’s storage saw emergence of new technologies and many developments happened during this period. The reason there are constant changes to data storage technology is due to the growth of data volume, variety and velocity. Video data in particular is the main component of overall capacity growth accoding to several studies. 

As storage costs continue to decline at a rate of 25% to 40% annually, more and more applications will become economic, further increasing demand for storage. This is becoming a self-reinforcing cycle.

## Structured vs Unstructured Data

One important concept about data to clarify is that there are 2 main categories of data formats organisations typically deal with. These two types are: 
- Structured data
- Unstructured data

Let's take a brief look at each:

### Structured Data

> _Structured data_ refers to data that can be organised and stored in tables composed of rows and columns

A typical example is data stored in an Excel spreadsheet or a database table containing details about the passengers of an airline company. Traditionally, most data companies stored and analysed was structured.This data is usually stored in a database such as Postgres or MySQL. Although structured data still exists today, it's becoming an increasingly smaller portion of the overall data being generated.

Below is a typical example of what structured data looks like. This is a table containig information about Students. Note that:
- Each _row_ represents a different student
- Each _column_ represents a particular detail about that student (such as their Last Name)

### Unstructured Data

> _Unstructured data_ refers data that is non-relational in nature and cannot be stored in tables of rows and columns

This is the more recent type of data that has exploded in growth after the introduction of the Internet, social media and smart devices. Streaming text data, video and images are typical examples of unstructured data. Traditional databases were not designed to store and analyse this type of data, and accordingly newer technologies were introduced to be able to handle the different data varieties and their increasing volumes.

The most popular tools to store and manipulate unstructured data are:

- Data lakes (such as Hadoop's HDFS)
- NoSQL data stores (such as HBase)

## Traditional Approaches to Data Storage

> Twenty years ago,  _storage silos_ were the essence of data storage global companies used. Applications were connected to the server operating system and the storage disks they ran on, upgrades meant expensive new hardware and possibly risky migrations, and the need to handle usage spikes meant the infrastructure was often over configured.

The variety of data formats and recording medium types has served as the criteria behind different approaches to storing data. From personal data storage devices to massive data centers and cloud-based data repositories, data engineers now have many options to organise and store corporate data.  

Traditionally, before the rise of big (mainly unstructured) data, organisations had the following data storage options to chose from:
- Direct-attached storage (DAS)
- Network-attached storage (NAS)
- Storage area networks (SAN)
- Software defined stored (SDS)
- Hyperconverged storage (HCS)

Let's look at each in more detail:

### 1. Direct-Attached Storage (DAS)

> Direct-attached storage stands for all types of physical data storage devices you can connect to a _single_ computer/server

Portable and affordable — yet only accessible by one computer at a time — DAS is a standard solution for keeping small-scale records data backups or for transferring data between devices.  

Popular types of direct-attached storage include:
- External hard drives or solid-state drives (SSD)
- Flash drives (USB sticks) 
- CD/DVD disks and other older methods (such as floppy disks)

### 2. Network-Attached Storage (NAS)  

> Network-attached storage is a special hardware unit, featuring a file-level architecture that can be accessed by more than one device as long as all users are connected to the same internal network

In essence, a NAS unit features several storage disks or hard drives, along with software for managing access requests. NAS units are easily to set up and can be used for both data storage and local file sharing

### 3. Storage Area Networks (SAN)

> Storage area networks (SANs) help assemble an even more complex on-premise data management architecture consisting of a dedicated network and a data storage system

SAN architecture features two components:

- A dedicated _network_ for data exchanges with network switches and load balancing
- Data _storage system_, consisting of on-premises hardware

The purpose of SAN is to act as a separate “highway” for transmitting data between servers and storage devices across an organisation. For instance, we can set up a dedicated low-latency data exchange lane between a server running big data analytics workloads and a storage system (i.e., data warehouse) hosting the processed data. Doing so helps prevent bottlenecks and delays for other users on the network. This approach can be considered a predecessor to cloud computing.

### 4. Software-Defined Storage (SDS)  

> A software-defined storage (SDS) system is a hardware-independent, software-based virtualisation storage architecture that can be used on any computing hardware platform

While NAS and SAN storage systems usually require vendor-supplied operating systems and supporting software, SDS lets users bring their own license to any type of server as it is a form of software virtualisation.

The two major benefits of SDS include:

- Scalability: You can reconfigure available hardware devices to better accommodate the required data
- Lower total cost of ownership (TCO): Instead of purchasing another NAS unit, you can easily re-deploy an existing one

To learn more about SDS, read the following blog:
-   [What is SDS](https://www.redhat.com/en/topics/data-storage/software-defined-storage)?



### 5. Hyperconverged Storage (HCS)  

> Hyperconverged storage allows you to virtualise on-premises storage resources to create a shared storage pool

Each hardware unit (node) gets virtualised and then connected in a cluster, which you can then manage as a unified system. The main advantage of HCS is hardware consolidation to leverage unused resources. Typically, 80% of data storage spending is tied to hardware, thus a company can significantly reduce maintenance costs by using HCS.

## Modern Approaches to Data Storage

More recently, three main data storage trends have gained momentum:

1. Structured data storage (using databases and data warehouses)
2. Unstructured data storage (using data lakes and NoSQL data stores)
3. Cloud storage

We'll examine each of these in more detail in the following sections.

### Storing Structured Data: Database and Data Warehouse Storage

> Database storage refers to tools leveraged to store _structured_ data in tables of rows and columns

> Data warehouses are used mainly to perform fast data analysis on company-wide _structured_ data. It's a more specialised tool providing complex features that a regular database does not provide.

In the past, data was organised and stored in file formats, which made them difficult to integrate and analyse. Databases were then introduced to enable organisations to connect the various data sets together to obtain a wider view. Structured data conforms to a tabular format of rows and columns, with relationship between the different rows and columns. 

Common examples of structured data are:
- Excel files
- Database tables 

#### Database Storage
- In databases, data is stored into tables of rows and columns
- The data itself is physically stored as one large file (or a number of smaller files) usually on a single machine
- Databases can store high-value as well as low-value data
- To scale up or down, data engineers must configure the properties controlling the database which that machine uses
- Recently, cloud-based relational databases have become common. Some examples include Amazon RDS and Oracle.

#### Data Warehouse Storage
- Data warehouses mainly store high-value, structured data that is produced after data transformation operations are completed
- The data usually comes from multiple other systems (which could be different company databases)
- The processed data is kept in the data warehouse for permanent, long-term storage and to enable analytics
- Traditionally, data warehouses were hosted on internal company hardware. More recently, they are being deployed in the cloud.

### Storing Complex Data Formats: Data Lakes and NoSQL

> More recently, unstructured high-volume big data has become around 80% of the data produced globally. This led to new tools being introduced to handle such an increasing volume and changing variety of data formats that relational database systems were longer able to cope with. 

In the past, data was mainly structured - meaning it was homogeneuous in the sense that it was all just rows and columns. More recently however, data formats like images and videos have become more complex and more unstructured in nature. Introducing this variety of data types means that data is now more heterogenous. We also started seeing semi-structured data, which is complex data type that is not relational but still has some structure to its format (such as JSON files). To be able to store, transform, integrate and analyse this data, organisations started using the more modern tools that started to be introduced in the late 2000's.

The 2 most popular types of complex data storage mechanisms currently used by corporations are: Data lakes and NoSQL data stores. 

Let's take a brief look at each:

#### Data lakes
- A _data lake_ is a centralised repository which leverages affordable commodity hardware. It enables companies to store all of its structured and unstructured data easily and at any scale
- One of the most popular data lake tools is Hadoop's distributed file system (HDFS)  
    - HDFS provides a cost-effective way of storing raw and processed data. It integrates well with other data technologies like databases, big data and cloud-based tools. 
    - HDFS also offers flexibility regarding its deployment, and can be configured either on-premise or in the cloud

#### NoSQL data stores
- NoSQL is an acronym for _Not Only SQL_. This term refers to a set of data management tools that can capture and store non-relational data.
- NoSQL tools provide flexible schemas that can store a variety of data types such as documents, key-value, columnar and graph data
- The most popular types of NoSQL data stores currently used in industry include MongoDB, Cassandra, HBase and ElasticSearch

## Cloud Data Storage

> Refers to the on-demand availability of computer system resources, especially data storage and computing power, without active management by the user

A growing trend in the field of data engineering is cloud computing. The main reason for its increasing popularity is that companies can get access to software or hardware needed on-demand, and only pay for actual use. This eliminates the hassle and costs of buying and managing on-premise data storage infrastructure such as servers. 

Many cloud vendors, such as Amazon AWS and Microsoft Azure, offer complementary services designed to help collect, manage, secure and analyse data at massive scale. Cloud providers use a distributed data storage model whereby the physical storage devices are geographically spread out across mulitple locations.

## Cloud Computing Architectures

Not all clouds are the same and not one type of cloud computing is best for all companies. Several different models, types, and services have evolved to help offer the right solution for every business need. 

There are three different ways to deploy cloud services: on a public cloud, private cloud, or hybrid cloud. 

### 1. Public cloud
- Public clouds are owned and operated by a third-party cloud service providers, which deliver their computing resources, like servers and storage, over the Internet. 
- Amazon AWS and Microsoft Azure are examples of public clouds
- Companies access these services and manage the corporate account using a web browser and an API

### 2. Private cloud
- Private clouds refers to cloud computing resources used exclusively by a single business or organisation. All services and infrastructure are invisible to anyone outside the organisation.
- A private cloud can be physically located on the company’s on-site datacenter, or virtually on a third-party service provider's infrastructure

### 3. Hybrid cloud
- Hybrid clouds combine public and private clouds, bound together by technology that allows data and applications to be shared between them
- By allowing data and applications to move between private and public clouds, a hybrid cloud gives businesses greater flexibility, more deployment options, and helps optimise existing infrastructure, security, and compliance
- This allows companies to store confidential data on in-house (private) hardware, while using the public infrastructure to host low-value data and related software

## Types of Cloud Storage

> There are three types of cloud data storage: object storage, file storage, and block storage. Each offers their own advantages and have their own use cases.

### 1. Object storage
- Treats each piece of data as a seperate object, keeping in in a seperate storage location and bundles it with related metadata and a unqiue ID to enable easy access and retrieval
- One example is Amazon S3, which is ideal for building modern applications that require scale and flexibility. S3 can also be used to import existing data stores for analytics or archiving.

### 2. File System Storage
- Views data in a file-hierarchy structure
- Some applications need to access shared files and require a file system. This type of storage is often supported with a Network Attached Storage (NAS) server. 
- One example is Amazon Elastic File System (EFS), which is ideal for use cases like large content repositories, development environments, and media stores.

### 3. Block Storage
- Treats data as blocks stored in sectors and tracks
- Other enterprise applications like databases often require dedicated, low latency storage for each host
- This is analagous to direct-attached storage (DAS) or a Storage Area Network (SAN). 
- One example is Oracle Cloud Storage (OCS), which offers the low latency required for high performance workloads

## Benefits of Cloud Storage

### Total Cost of Ownership
- With cloud storage, there is no hardware to purchase, storage to provision, or specialised hardware experts to hire
- Companies can add or remove capacity on demand, quickly change performance and retention characteristics, and only pay for storage that are actually used
- Less frequently accessed data can even be automatically moved to lower cost tiers to help save costs

### Time to Deployment
- Cloud storage allows data engineers to quickly deliver the exact amount of storage needed, right when it's needed
- Cloud storage deployment times are usually very fast and require only a few minutes

### Performance
- The biggest cloud computing services run on a worldwide network of secure datacenters, which are regularly upgraded to the latest generation of fast and efficient computing hardware
- This offers several benefits over a single corporate datacenter, including reduced network latency for applications and greater economies of scale

### Reliability
- Cloud computing makes data backup, disaster recovery, and business continuity easier and less expensive because data can be mirrored at multiple redundant sites
- If one server goes down, organisations can access their files from a back-up server. This is called highly-available infrastructure, meaning that the uptime is almost guaranteed at nearly 100%.

### Security 
- Many cloud providers offer a broad set of policies, technologies, and controls that strengthen overall security 
- Cloud companies often rely on robust cybersecurity measures and they hire 3rd party cybersecurity experts to protect sensitive data
- Files stored on cloud servers are usually encrypted to help enhance security of the data
- Cloud providers rely on firewalls, too, to help protect sensitive files. As the name suggests, this technology acts a like a wall keeping data safe.

It's worth noting that cloud computing supports handling structured data in addition to complex, unstructured data types.

## Key Takeaways

- There is an increasing quantity and variety of data being generated by global companies. Most of this new data is complex in nature and is usually machine generated.
- The 2 main types of data being generated and stored are: structured (rows and columns) and unstructured (such as text, video and image data)
- Traditionally, there were 5 main types of data storage approaches used: direct attached storage, network attached storage, storage area networks, software defined storage and hyper-converged storage
- To store structured data, the most common types of storage include: databases and data warehouse systems. Although both are similar, we should note that a data warehouse is a more specialised tool that usually integrate data from multiple databases.
- To store complex data, data lakes and NoSQL data stores are the 2 most common approaches used. Complex data is usually unstructured, but could also be structured or semi-structured.
- Data lakes are a modern approach to store complex data structures. They are a centralised storage system that relies on commodity hardware. Hadoop's HDFS is one of the most popular data lakes used in industry.
- NoSQL data stores are a family of tools designed to store and interact with different data types such as text and graph data. They don't all handle data the same way.
- The main trend nowadays for data storage in industry is to migrate to cloud-based solutions, rather than use on-premise hardware (which is more expensive)
- Cloud-based architectures can be: public (shared), private (exclusive to a specific company) or a hybrid of both. The exact model to deploy depends on the business and technical requirements of the organisation.
- Cloud technologies offer companies numerous benefits such as reduced costs, faster deployment times, easier scaling, enhanced security and the ability to leverage automated data manipulation tools.
